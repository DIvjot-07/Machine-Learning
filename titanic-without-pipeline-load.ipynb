```python
import pickle
import numpy as np
```

```python
ohe_sex = pickle.load(open('/kaggle/input/notebooks/divjotbedi/titanic-without-pipeline/models/oe_sex.pkl', 'rb'))
ohe_embarked = pickle.load(open('/kaggle/input/notebooks/divjotbedi/titanic-without-pipeline/models/oe_embarked.pkl', 'rb'))
clf = pickle.load(open('/kaggle/input/notebooks/divjotbedi/titanic-without-pipeline/models/clf.pkl', 'rb'))
```

```python
input=np.array([[1,'male',36.0,0,0,75.2417,'C']],dtype=object).reshape(1,7)
```

```python
input_sex=ohe_sex.transform(input[:,1].reshape(1,1))
```

```python
input_sex
```

```python
input_embarked=ohe_embarked.transform(input[:,-1].reshape(1,1))
```

```python
input_embarked
```

```python
input_age=input[:,2].reshape(1,1)
```

```python
input_age
```

```python
input_transformed=np.concatenate((input[:,[0,3,4,5]],input_sex,input_embarked,input_age),axis=1)
```
