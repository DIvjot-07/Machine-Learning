```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder ,MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split 
```

```python
df=pd.read_csv('/kaggle/input/datasets/brendan45774/test-file/tested.csv')
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
train_X,test_X,train_y,test_y=train_test_split(
    df.drop('Survived',axis=1),
df['Survived'],
    test_size=0.3,
    random_state=42)
```

```python
train_X.head()
```

```python
test_X.head()
```

```python
train_y.head()
```

```python
df.isnull().sum()
```

```python
si=SimpleImputer()
train_X_age=si.fit_transform(train_X[['Age']])
test_X_age=si.fit_transform(test_X[['Age']])
train_X_fare=si.fit_transform(train_X[['Fare']])
test_X_fare=si.fit_transform(test_X[['Fare']])
```

```python
pd.DataFrame(train_X_age).isnull().sum()
```

```python
pd.DataFrame(train_X_fare).isnull().sum()
```

```python
oe_sex=OneHotEncoder(sparse_output=False)
train_X_sex=oe_sex.fit_transform(train_X[['Sex']])
test_X_sex=oe_sex.fit_transform(test_X[['Sex']])
```

```python
oe_embarked=OneHotEncoder(sparse_output=False)
train_X_embarked=oe_embarked.fit_transform(train_X[['Embarked']])
test_X_embarked=oe_embarked.fit_transform(test_X[['Embarked']])
```

```python
train_X_embarked
```

```python
test_X_sex
```

```python
train_X_rem=train_X.drop(columns=['Age','Sex','Embarked','Fare'])
test_X_rem=test_X.drop(columns=['Age','Sex','Embarked','Fare'])
```

```python
import numpy as np
train_X_transform=np.concatenate((train_X_rem,train_X_fare,train_X_embarked,train_X_sex,train_X_age),axis=1)
test_X_transform=np.concatenate((test_X_rem,test_X_fare,test_X_embarked,test_X_sex,test_X_age),axis=1)
```

```python
train_X_transform
```

```python
train_X_transform.shape
```

```python
train_X_rem.head()
```

```python
df_train_transformed=pd.DataFrame(train_X_transform,columns=['Pclass','SibSp','Parch','Fare','C','Q','S','Female','Male','Age'])
```

```python
df_train_transformed.head()
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(train_X_transform,train_y)
```

```python
pred_y=clf.predict(test_X_transform)
pred_y
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(pred_y,test_y)
```
